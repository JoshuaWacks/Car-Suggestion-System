```python
%matplotlib inline

import pandas as pd
import cv2
from matplotlib import pyplot as plt
import torch
import numpy as np
import os
import xml.etree.ElementTree as ET
from torchvision.models.detection import fasterrcnn_resnet50_fpn_v2, FasterRCNN_ResNet50_FPN_V2_Weights
from torchvision.io.image import decode_image
from torchvision.utils import draw_bounding_boxes
from torchvision.transforms.functional import to_pil_image

torch.cuda.is_available()
torch.cuda.device_count()
```

```python
weights = FasterRCNN_ResNet50_FPN_V2_Weights.DEFAULT
model = fasterrcnn_resnet50_fpn_v2(weights=weights, box_score_thresh=0.6)
model.eval()
```

```python
preprocess = weights.transforms()
```

```python
for file in os.listdir(parent_folder):
    if file.endswith('xml'):
        continue
    full_image_path = os.path.join(parent_folder,file)
    image = decode_image(full_image_path)
    batch = [preprocess(image)]
    prediction = model(batch)[0]

    main_car = []
    for boxes,labels in zip(prediction['boxes'],prediction['labels']):
        if labels == 3:
            main_car = boxes.reshape(1,-1)
        break
    
    print(main_car,prediction['scores'][0])
    # labels = [weights.meta["categories"][i] for i in prediction["labels"]]
    if len(main_car) == 0:
        image = cv2.imread(full_image_path)
        cv2.imshow('Image', image) 
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        continue
    box = draw_bounding_boxes(image, boxes=main_car,
                            # labels=labels,
                            colors="red",
                            width=4, font_size=30)
    im = to_pil_image(box.detach())
    im.show()

    plt.show()
```
