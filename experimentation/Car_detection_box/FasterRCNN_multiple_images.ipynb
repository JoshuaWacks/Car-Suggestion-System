```python
%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import torch
import os
from torchvision.models.detection import fasterrcnn_resnet50_fpn_v2, FasterRCNN_ResNet50_FPN_V2_Weights
from torchvision.io.image import decode_image
from torchvision.utils import draw_bounding_boxes
from torchvision.transforms.functional import to_pil_image


torch.cuda.is_available()
torch.cuda.device_count()
```

```python
weights = FasterRCNN_ResNet50_FPN_V2_Weights.DEFAULT
model = fasterrcnn_resnet50_fpn_v2(weights=weights, box_score_thresh=0.7)
model.eval()
```

```python
# Step 2: Initialize the inference transforms
preprocess = weights.transforms()
```

```python
main_folder = '../../data/Car Labelling/Cars Dataset/train/'
for car_type in os.listdir(main_folder):
    spec_car_folder = os.path.join(main_folder,car_type)
    for i,car in enumerate(os.listdir(spec_car_folder)):
        file_path = os.path.join(spec_car_folder, car)
        
        image = decode_image(file_path)
        batch = [preprocess(image)]
        prediction = model(batch)[0]
        labels = [weights.meta["categories"][i] for i in prediction["labels"]]
        box = draw_bounding_boxes(image, boxes=prediction["boxes"],
                                labels=labels,
                                colors="red",
                                width=4, font_size=30)
        im = to_pil_image(box.detach())
        im.show()

        plt.show()
        if i > 10:
            break
```
